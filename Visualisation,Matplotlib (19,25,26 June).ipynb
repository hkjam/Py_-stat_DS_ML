```python
##Bar chart
data={"English":45,"Math":56,"GS":59,"Social Science":89,"Hindi":51}
Subject=data.keys()
Marks=data.values()

## plot on bar graph
plt.bar(Subject,Marks) ##subject on x axis and Marks on y 
##color
plt.bar(Subject,Marks,color=['g','k','r','b','y'])
##Asthetics
plt.title("This is my bar graph")
plt.ylabel("Marks")
plt.xlabel("Subjects")
##plt.show()
```

```python
##to increse the size
fig=plt.figure(figsize=(10,15))
#plot
plt.bar(Subject,Marks,color="m")

##labels
plt.title("Marks of Ankit in different Subject")
plt.xlabel("Subjects")
plt.ylabel("Marks")
```

```python
##To interchange axis i.e Subject on y axis and Marks on x axis. Subject and marks should be made list.

Subject=list(data.keys())
Marks=list(data.values())
plt.barh(Subject,Marks,color=['g','k','r','b','y'])## label gets interchange
```

```python
## Histogram: used to plot distribution
## It is a continuous graph ploting continuous data 
## it has data between max and min value
## it has no gap in between as in bar graph
```

```python
##Steps:1. take range of data
#2. Ascending order
#3.Bins are created 
#4.Plot
```

```python
import matplotlib.pyplot as plt
import numpy as np

## Data
data=np.random.normal(100,20,300) ## 100 is mean, 20 is S.D and 300 is no. of points
##plot
plt.hist(data)
plt.show()
```

```python
plt.hist(data,bins=50)## default bin is 10
plt.show()
```

```python
plt.hist(data,bins=50,orientation="horizontal")
plt.show()
```

```python
plt.hist(data,bins=50,orientation="horizontal",color='g')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = 'English', 'Math', 'Hindi', 'SS'
sizes = [85, 60, 75, 80]
explode = (0, 0.1, 0, 0.1)  # only "explode" the 2nd and 4th slice (i.e. 'Math',"ss")

fig1, ax1 = plt.subplots() ## fig creating subplot,,ax1 is location where we plot pie chart

ax1.pie(sizes, explode=explode, labels=labels,autopct='%2.4f%%',
        shadow= True, startangle=0)

##ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
plt.savefig("piechart.png")
```
